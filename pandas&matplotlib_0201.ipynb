#エクササイズ - matplotlib

```
import numpy as np
import matplotlib.pyplot as plt
```

```
#mat_ex06.py
x = np.linspace(-5, 5, 100)
y = x ** 2
plt.plot(x, y)
plt.grid()
plt.show()
```

```
#mat_ex07.py

math = np.array([60, 70, 66, 88, 82, 75, 90, 72, 48, 52])
science = np.array([66, 64, 76, 92, 90, 75, 84, 70, 40, 48])

plt.scatter(math, science)
plt.grid()
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.show()
```

```
#mat_ex08.py

sales = np.array([5000, 3000, 2000])
sales_label = np.array(["A", "B", "C"])

figure = plt.figure()

axes1 = figure.add_subplot(2, 2, 1)
axes1.bar(sales_label, sales)

axes2 = figure.add_subplot(1, 2, 2)
axes2.pie(sales, labels=sales_label, counterclock=False, startangle=90)

plt.show()
```

#pandas - part3 - CSVファイルの処理 - Exercise

```
import pandas as pd

pd.options.display.precision = 2
pd.options.display.float_format = '{:.2f}'.format

population = pd.read_csv("population.csv", index_col=0)
population
```

```
#3.1 2010年以降のデータをCSVファイルに出力してください。ファイル名はoutput1.csvとします。
population[population.index >= 2010].to_csv("output1.csv")
```

```
#3.2 total列（maleとfamaleの合計）とdiff列（maleとfemaleの差分）を追加してCSVファイルに出力してください。ファイル名はoutput2.csvとします。

total = population["male"] + population["female"]
#cool = pd.Series(total, "total")

minus = population["male"] - population["female"]

population["total"] = total
population["diff"] = minus
population.to_csv("output2.csv")
```

```
#3.2 total列（maleとfamaleの合計）とdiff列（maleとfemaleの差分）を追加してJSONファイルに出力してください。ファイル名はoutput3.jsonとします。

total = population["male"] + population["female"]
#cool = pd.Series(total, "total")

minus = population["male"] - population["female"]

population["total"] = total
population["diff"] = minus
population.to_json("output3.json", indent=2)
```
